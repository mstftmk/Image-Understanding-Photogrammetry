```python
import cv2

import numpy as np
import matplotlib
from matplotlib import pyplot as plt

#!matplotlib inline
```

```python
cv2.__version__
```

```python
img = cv2.imread('robotech.jpg')
img.shape
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
ksize = (15, 15)
blur_img = cv2.blur(img, ksize)
gblur_img = cv2.GaussianBlur(img, ksize, 4.0)

fig, axs = plt.subplots(1, 2, figsize=(12,8))
axs[0].imshow(blur_img)
axs[1].imshow(gblur_img)
```

```python
new_img = np.zeros((128, 256), np.uint8)
new_img[32:96, 96:164] = 255
```

```python
cv2.imwrite('test_image.png', new_img)
```
